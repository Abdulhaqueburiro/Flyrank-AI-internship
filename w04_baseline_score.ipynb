## 1. My rule and its reason codes

*Write the rule in plain words first. Then the reason codes it can output.*

```python
rule = """
If a page has high impressions but a low CTR, prioritize it for optimization because it has the greatest potential to gain more clicks.
"""

reason_codes = {
    "LOW_CTR_HIGH_IMPRESSIONS": "High impressions but CTR is below expected.",
    "STALE_CONTENT": "Content has not been updated recently."
}

print(rule)
print(reason_codes)
```

## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

```python
import pandas as pd

# Sample data
data = {
    "page": [
        "Home",
        "Projects",
        "Blog",
        "About",
        "Contact",
        "Portfolio",
        "Services",
        "ML Project 1",
        "ML Project 2",
        "Resume"
    ],
    "impressions": [1200, 950, 1800, 500, 300, 1500, 400, 1300, 900, 700],
    "ctr": [1.2, 2.5, 0.8, 3.5, 1.0, 1.8, 2.0, 0.9, 1.5, 2.8]
}

df = pd.DataFrame(data)

# Score calculation
df["score"] = df["impressions"] * (3 - df["ctr"])

# Reason code
df["reason_code"] = "LOW_CTR_HIGH_IMPRESSIONS"

# Action label
df["action"] = "Optimize Title & Meta Description"

# Rank by score
df = df.sort_values("score", ascending=False)

# Save CSV
df.to_csv("baseline_action_score.csv", index=False)

print(df.head(10))
print("\nCSV saved successfully as baseline_action_score.csv")
```
