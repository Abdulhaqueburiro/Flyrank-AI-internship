## 2. Intended use and limits

*Who uses this, for what — and where it stops being valid.*

```python
intended_use = """
Intended Use:
This playbook is designed to help content and SEO teams prioritize pages that need optimization.
It provides ranked recommendations based on observed signals such as impressions, CTR, and content freshness.

Limits:
- This is a decision-support tool, not an automated publishing system.
- Human review is required before making content changes.
- Results depend on the quality and freshness of the available data.
- Recommendations may not generalize to every website or industry.
"""

print(intended_use)
```

## 3. Human review + the no-go list

*What a person must check before acting. What should never be automated.*

```python
human_review = """
Human Review Rules:
- Review all recommendations before taking action.
- Verify that the page is still relevant.
- Check business goals before updating content.
- Confirm that recommendations align with current SEO strategy.

No-Go List:
- Do not automatically publish content.
- Do not delete pages based only on model output.
- Do not make major SEO decisions without human approval.
- Do not use the playbook as the only decision-making tool.
"""

print(human_review)
```

## 4. Monitoring / retrain triggers

*What would tell you the recommendations went stale?*

```python
monitoring = """
Monitoring:
- Monitor CTR, impressions, and ranking trends weekly.
- Review model performance every month.
- Compare new data with previous results.

Retrain Triggers:
- Significant drop in prediction quality.
- Major changes in search behavior.
- Large updates to the dataset.
- New SEO strategy or website structure changes.
"""

print(monitoring)
```
